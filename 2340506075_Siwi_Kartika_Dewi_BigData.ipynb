```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

dataset = pd.read_csv('/content/Warehouse_and_Retail_Sales.csv')

# Daftar kolom numerik yang akan dinormalisasi
numerical_columns = ["YEAR", "MONTH", "RETAIL SALES", "RETAIL TRANSFERS", "WAREHOUSE SALES"]

# Inisialisasi MinMaxScaler
scaler = MinMaxScaler()

# Normalisasi hanya kolom numerik
dataset[numerical_columns] = scaler.fit_transform(dataset[numerical_columns])

# Menampilkan dataset setelah normalisasi
print("\nSetelah Normalisasi:")
print(dataset.head(10))
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

dataset = pd.read_csv('/content/Warehouse_and_Retail_Sales.csv')

# Daftar kolom numerik yang akan distandarisasi
numerical_columns = ["YEAR", "MONTH", "RETAIL SALES", "RETAIL TRANSFERS", "WAREHOUSE SALES"]

# Inisialisasi StandardScaler
scaler = StandardScaler()

# Standarisasi hanya kolom numerik
dataset[numerical_columns] = scaler.fit_transform(dataset[numerical_columns])

# Menampilkan dataset setelah standarisasi
print("\nSetelah Standarisasi:")
print(dataset.head(10))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Membaca dataset
dataset = pd.read_csv('/content/Warehouse_and_Retail_Sales.csv')

# Mengambil 10 pemasok teratas berdasarkan total penjualan retail
top_suppliers = dataset.groupby('SUPPLIER')['RETAIL SALES'].sum().nlargest(10).reset_index()

# Diagram 1: Scatter Plot dengan Seaborn
plt.figure(figsize=(12, 8))  # Ukuran grafik
sns.scatterplot(x='SUPPLIER', y='RETAIL SALES', data=top_suppliers, s=100, color='fuchsia')

# Menambahkan label dan judul
plt.title('Scatter Plot: Top 10 Retail Sales per Supplier (Seaborn)', fontsize=16)
plt.xlabel('Supplier', fontsize=14)
plt.ylabel('Total Retail Sales', fontsize=14)

# Memutar label sumbu X agar lebih mudah dibaca
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Diagram 2: Grafik Batang dengan Matplotlib
# Ekstrak data untuk Matplotlib
suppliers = top_suppliers['SUPPLIER']
sales = top_suppliers['RETAIL SALES']

plt.figure(figsize=(20, 10))  # Ukuran grafik
plt.bar(suppliers, sales, color='pink')  # Membuat grafik batang

# Menambahkan label dan judul
plt.title('Bar Chart: Top 10 Retail Sales per Supplier (Matplotlib)', fontsize=16)
plt.xlabel('Supplier', fontsize=14)
plt.ylabel('Total Retail Sales', fontsize=14)

# Memutar label sumbu X agar lebih mudah dibaca
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
